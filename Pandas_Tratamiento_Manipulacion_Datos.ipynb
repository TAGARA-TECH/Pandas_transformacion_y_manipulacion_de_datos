```python
# Limpiar la columna 'descripcion_local' eliminando caracteres que no son alfanuméricos o guiones
# Corregir el patrón de la expresión regular para cerrar correctamente el set de caracteres
datos['descripcion_local'] = datos['descripcion_local'].str.replace('[^a-zA-Z0-9\-\-]', ' ', regex=True)

# Puedes verificar el resultado en un ejemplo
print(datos['descripcion_local'][3169])
```

```python
# borrar guiones que no tengan un caracter a la derecha o a la izquierda
# Usamos (?<!\w)-(?!\w) para encontrar guiones que no están rodeados por caracteres de palabra.
# Reemplazamos esos guiones con un espacio.
datos['descripcion_local'] = datos['descripcion_local'].str.replace(r'(?<!\w)-(?!\w)', ' ', regex=True)

datos['descripcion_local'].head()
```
